---

Vamos entender aqui que tipo visualizações são as mais adequadas para responder a perguntas sobre os dados e como podemos produzir essas visualizações de forma simples em R. Nosso foco, aqui, será na produção de gráficos de Exploração (em oposição aos gráficos de Apresentação).

```python
%load_ext rpy2.ipython
```

```python
%%R

set.seed(123)

x = seq(1,100,1)
y = runif(100)*10
z = runif(100)*3
t = runif(100)*5
```

```python
%%R

y
```

```python
%%R

plot(x, y, type='l', main='Plot', col='red', xlab = 'valores de x', ylab ='valores de y e sqrt(y)', cex.main=2)
# cex.lab for axis labels and cex.axis for axis text
lines(x, sqrt(y), type='l', col='blue')
legend("topleft",c("y","sqrt(y)"),fill=c("red","blue"))
lines(c(1,100), c(mean(y),mean(y)), type='l', col='black', lty=2)
text(50, mean(y)+1, round(mean(y),2), cex=1.5)

```

# Outros Gráficos

```python
%%R

hist(y, breaks = 30, main='Histogram' , col='green' )

plot(x, y, main='Scatter plot')

boxplot(y,z,t,names = c("y", "z", "t"), col = c("orange","red","blue"), main='Boxplot' )

barplot(c(mean(y), mean(z), mean(t)),names = c("y", "z", "t"), col='blue',density=10 , main='Bar plot', ylim=c(0,6))
text(c(1,2,3), c(mean(y)+0.5, mean(z)+0.5, mean(t)+0.5), labels=c(round(mean(y),2), round(mean(z),2), round(mean(t),2)), cex=1.5)

pie(c(mean(y), mean(z), mean(t)), col=c("orange","red","blue") ,density=80, main='Pie plot' )
```
