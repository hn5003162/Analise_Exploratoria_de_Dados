```python
# 3) Qual é a região com maior perda de alimentos?
# mudar forma de cálculo usando groupby() // 22/09/2023 Hideki

#Cria um datafame somente das perdar por região. Elimina as regiões que não tem dados (NA)
base_perda = base[["nome_regiao","area_plantada","perdas"]]
# Elimina todos os registro que tenham NA no campo perdas
base_perda = base_perda.dropna(subset=['perdas'])

#Area plantada em números absolutos de 2010 a 2017"
print()
print("Area plantada:")
print("Norte       : ",base_perda[base_perda.nome_regiao == "Norte"].area_plantada.sum())
print("Nordeste    : ",base_perda[base_perda.nome_regiao == "Nordeste"].area_plantada.sum())
print("Centro-oeste: ",base_perda[base_perda.nome_regiao == "Centro-oeste"].area_plantada.sum())
print("Sudeste     : ",base_perda[base_perda.nome_regiao == "Sudeste"].area_plantada.sum())
print("Sul         : ",base_perda[base_perda.nome_regiao == "Sul"].area_plantada.sum())



#Perdas em números absolutos:
print()
print("Perdas de 2010 a 2017:")
print("Norte       :",base_perda[base_perda.nome_regiao == "Norte"].perdas.sum())
print("Nordeste    :",base_perda[base_perda.nome_regiao == "Nordeste"].perdas.sum())
print("Centro-oeste:",base_perda[base_perda.nome_regiao == "Centro-oeste"].perdas.sum())
print("Sudeste     :",base_perda[base_perda.nome_regiao == "Sudeste"].perdas.sum())
print("Sul         :",base_perda[base_perda.nome_regiao == "Sul"].perdas.sum())

#Perdas proporcionais à area plantada
print()
print("Perdas proporcionais à area plantada de 2010 a 2017:")
print("Norte       : ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Norte"].perdas.sum() / base_perda[base_perda.nome_regiao == "Norte"].area_plantada.sum()))
print("Nordeste    : ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Nordeste"].perdas.sum() / base_perda[base_perda.nome_regiao == "Nordeste"].area_plantada.sum()))
print("Centro-oeste: ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Centro-oeste"].perdas.sum() / base_perda[base_perda.nome_regiao == "Centro-oeste"].area_plantada.sum()))
print("Sudeste     : ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Sudeste"].perdas.sum() / base_perda[base_perda.nome_regiao == "Sudeste"].area_plantada.sum()))
print("Sul         : ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Sul"].perdas.sum() / base_perda[base_perda.nome_regiao == "Sul"].area_plantada.sum()))

```

```python
# 1) Qual é a relação entre o plantio e a emissão de gases de efeito estufa?
base_plantada_emissao = base[["area_plantada","emissao"]]
#Retira todo e qualquer registro com NA, seja da area plantada ou na emissao"
base_plantada_emissao = base_plantada_emissao.dropna()
print("Emissão de gases por hectare:",base_plantada_emissao.emissao.sum()/base_plantada_emissao.area_plantada.sum())





```

2) Qual é a perda de cada região por ano?


```python
# Versão Hideki usando o campo perdas do dataframe


base_perda_regiao = base[["ano","nome_regiao","perdas"]]
base_perda_regiao.dropna()


base_perda_regiao = base_perda_regiao.groupby(['ano', 'nome_regiao'])

base_perda_regiao = base_perda_regiao.sum()


display(base_perda_regiao)


```

```python
# Agrupado por região e ano

base_perda_regiao = base[["ano","nome_regiao","perdas"]]
base_perda_regiao.dropna()

base_perda_regiao = base_perda_regiao.groupby(['nome_regiao','ano'])
base_perda_regiao = base_perda_regiao.sum()

display(base_perda_regiao)

#Perdas totais por ano
base_perda_regiao = base[["ano","nome_regiao","perdas"]]
base_perda_regiao.dropna()

base_perda_regiao = base_perda_regiao.groupby(['ano'])
base_perda_regiao = base_perda_regiao.sum()

display(base_perda_regiao)

```

```python
# Qual é a relação de perdas e emissão?

# 1) Qual é a relação entre o plantio e a emissão de gases de efeito estufa?
base_perdas_emissao = base[["nome_regiao","ano","area_plantada","perdas","emissao"]]
#Retira todo e qualquer registro com NA, seja da area plantada ou na emissao"
base_perdas_emissao.dropna()

base_perdas_emissao = base_perdas_emissao.groupby(['nome_regiao','ano'])
base_perdas_emissao = base_perdas_emissao.sum()

display(base_perdas_emissao)



```
