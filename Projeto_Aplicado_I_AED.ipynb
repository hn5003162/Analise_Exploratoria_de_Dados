```python
# 3) Qual é a região com maior perda de alimentos?


#Cria um datafame somente das perdar por região. Elimina as regiões que não tem dados (NA)
base_perda = base[["nome_regiao","area_plantada","perdas"]]
# Elimina todos os registro que tenham NA no campo perdas
base_perda = base_perda.dropna(subset=['perdas'])

#Area plantada em números absolutos de 2010 a 2017"
print()
print("Area plantada:")
print("Norte       : ",base_perda[base_perda.nome_regiao == "Norte"].area_plantada.sum())
print("Nordeste    : ",base_perda[base_perda.nome_regiao == "Nordeste"].area_plantada.sum())
print("Centro-oeste: ",base_perda[base_perda.nome_regiao == "Centro-oeste"].area_plantada.sum())
print("Sudeste     : ",base_perda[base_perda.nome_regiao == "Sudeste"].area_plantada.sum())
print("Sul         : ",base_perda[base_perda.nome_regiao == "Sul"].area_plantada.sum())



#Perdas em números absolutos:
print()
print("Perdas de 2010 a 2017:")
print("Norte       :",base_perda[base_perda.nome_regiao == "Norte"].perdas.sum())
print("Nordeste    :",base_perda[base_perda.nome_regiao == "Nordeste"].perdas.sum())
print("Centro-oeste:",base_perda[base_perda.nome_regiao == "Centro-oeste"].perdas.sum())
print("Sudeste     :",base_perda[base_perda.nome_regiao == "Sudeste"].perdas.sum())
print("Sul         :",base_perda[base_perda.nome_regiao == "Sul"].perdas.sum())

#Perdas proporcionais à area plantada
print()
print("Perdas proporcionais à area plantada de 2010 a 2017:")
print("Norte       : ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Norte"].perdas.sum() / base_perda[base_perda.nome_regiao == "Norte"].area_plantada.sum()))
print("Nordeste    : ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Nordeste"].perdas.sum() / base_perda[base_perda.nome_regiao == "Nordeste"].area_plantada.sum()))
print("Centro-oeste: ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Centro-oeste"].perdas.sum() / base_perda[base_perda.nome_regiao == "Centro-oeste"].area_plantada.sum()))
print("Sudeste     : ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Sudeste"].perdas.sum() / base_perda[base_perda.nome_regiao == "Sudeste"].area_plantada.sum()))
print("Sul         : ","{0:.5f}".format(base_perda[base_perda.nome_regiao == "Sul"].perdas.sum() / base_perda[base_perda.nome_regiao == "Sul"].area_plantada.sum()))

```

```python
# 1) Qual é a relação entre o plantio e a emissão de gases de efeito estufa?
base_plantada_emissao = base[["area_plantada","emissao"]]
#Retira todo e qualquer registro com NA, seja da area plantada ou na emissao"
base_plantada_emissao = base_plantada_emissao.dropna()
print("Emissão de gases por hectare:",base_plantada_emissao.emissao.sum()/base_plantada_emissao.area_plantada.sum())





```

2) Qual é a perda de alimentos na colheita de cada região?


```python
# Alterado por Natália

base_perda = base[["ano","nome_regiao","area_plantada", "area_colhida"]]
base_perda = base_perda.dropna()

base_perda = base_perda.groupby(['ano', 'nome_regiao'])
base_perda = base_perda.sum()

base_perda['perda'] = base_perda["area_plantada"].sub(base_perda["area_colhida"])

display(base_perda)
```

4) Qual a relação do desperdício com a emissão de gases do efeito estufa?

```python
# Agrupando por ano e região e somando as áreas plantadas, áreas colhidas e emissões
base_emissao = base[["ano", "nome_regiao", "area_plantada", "area_colhida", "emissao"]]
base_emissao = base_emissao.dropna()

base_emissao = base_emissao.groupby(['ano', 'nome_regiao']).sum()

base_emissao['perda'] = base_emissao["area_plantada"].sub(base_emissao["area_colhida"])

# Calculando a relação entre a perda e a emissao de gases de efeito estufa
base_emissao['relacao_perda_emissao'] = base_emissao['perda'] / base_emissao['emissao']
display(base_emissao)
```

4.1 Diminuir a perda reflete diretamente na diminuição da emissão dos gases?

```python
# Correlacionando a perda e a emissão de gases de efeito estufa
correlacao = base_emissao['perda'].corr(base_emissao['emissao'])

# Correlação negativa: significa que há uma relação inversa entre perda e emissão.
# Correlação positiva: significa que há uma relação direta entre perda e emissão.

print(f"A correlação entre a perda e emissão de gases de efeito estufa é: {correlacao}")
```

4.2 Qual a relação entre a produtividade e a emissão de gases do efeito estufa?

```python
# Calculo da produtividade
base['produtividade'] = base['producao_cafe'] / base['area_colhida']
# Agrupamento por ano e região e calculo da média da produtividade e a soma das emissões
base_prod_emissao = base.groupby(['ano', 'nome_regiao'])[['produtividade', 'emissao']].agg({'produtividade':'mean', 'emissao':'sum'})
# Relação entre a produtividade e emissão de gases de efeito estufa
base_prod_emissao['relacao_prod_emissao'] = base_prod_emissao['produtividade'] / base_prod_emissao['emissao']
display(base_prod_emissao)
```

5) Qual a relaçção entre a produtividade e a perda considerando os 5 estados com maior produtividade?


```python
# Calculo da produtividade
base['produtividade'] = base['producao_cafe'] / base['area_colhida']
# Agrupamento por estado e calculo da média da produtividade e a soma das perdas
base_prod_perda = base.groupby(['nome_uf'])[['produtividade', 'perdas']].agg({'produtividade':'mean', 'perdas':'sum'})
# Ordenamento dos estados pela produtividade em ordem decrescente e selecionando os 5 primeiros
top_5_estados = base_prod_perda.nlargest(5, 'produtividade')
# Relação entre produtividade e perda
top_5_estados['relacao_prod_perda'] = top_5_estados['produtividade'] / top_5_estados['perdas']
display(top_5_estados)
```
